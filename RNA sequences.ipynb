```python
import numpy as np
import scipy.stats

#Only works if assuming normal distribution
list1 = list(dict_sample4rec.values())
threshold1 = np.mean(list1)+ 2*np.std(list1)

list2 = list(dict_sample3rec.values())
threshold2 = np.mean(list2)+ 2*np.std(list2)


def mean_confidence_interval(data, confidence=0.95):
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
    return m, m-h, m+h

#print(mean_confidence_interval(list1))

enriched_seq_sample3 = []
for key in dict_sample3rec:
    if dict_sample3rec[key] > threshold2 :
        enriched_seq_sample3.append(key)
    
print(enriched_seq_sample3)
print(len(enriched_seq_sample3))


enriched_seq_sample4 = []
for key in dict_sample4rec:
    if dict_sample4rec[key] > threshold1 :
        enriched_seq_sample4.append(key)
    
print(enriched_seq_sample4)
print(len(enriched_seq_sample4))
```

```python
#Assuming the underlying distribution is unknown- use bootsrapping
#get the normal distribution of 0.975 quantile- then just take its mean to estimate the cutoff

print(len(list1))
#print(list1)
N = 10000

def threshold_function(data, confidence):
    quantile_up_estimates = []
    quantile_low_estimates = []
    for i in range(10000):
        bootstrap = random.choices(list(data), k = len(data))
        a = np.array(bootstrap)
        up = np.quantile(a, (1+confidence)/2)
        down = np.quantile(a, (1-confidence)/2)
        quantile_up_estimates.append(up)
        quantile_low_estimates.append(down)
    up_threshold = np.mean(quantile_up_estimates)
    low_threshold = np.mean(quantile_low_estimates)
    return low_threshold, up_threshold

thres_sample4 = threshold_function(list1, 0.95)
thres_sample3 = threshold_function(list2, 0.95)
print(thres_sample3[0])
```
