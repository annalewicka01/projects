## Modelling approaches
One of the approaches to mathematical modelling to epidemics can be based on partial differential equations. A crucial limitation of this strategy regarding the COVID-19 pandemic is that they present a static picture of the universe aiming to approach a state of permanent equilibrium. Factually, many complex systems in nature (such as societies) are constantly changing. Another approach to mathematical modelling, probabilistic cellular automata, is designed to capture these evolutionary dynamics. It 
features a grid of cells representing ‘individuals’ in an assigned state, which interact with their neighbouring cells. Over a given period, the states of the cells are updated following specified rules, simulating complex phenomena through evolving. 

## Our solution to the problem
To gain insight into the impact of some of the health policy measures on the Covid-19 pandemic and make predictions, we set our model as a probabilistic cellular automaton. Our project began with defining three primary states for individual cells: susceptible (S), infected (I) and recovered (R). With these established, the additional conditions of quarantined (Q) and vaccinated (V) were implemented. Subsequently, the project required a set of rules to be developed, according to which the states will be updated. These rules can then be adjusted to simulate various conditions throughout the pandemic.

# Model description

As a first step, we created a grid of cells which would act as of universe of individuals, where we can run our probabilistic simulations. We allowed the sources of outbreak to be spread randomly in the population.

Next, we established rules regarding the conditions of getting infected. We assumed a cell is neighbouring another cell when they are in Moore neighbourhood with each other. This illustrates what a Moore neighbourhood is (yellow cells are all Moore neighbours to the purple cell):

```python
moore=G3.Moore()
plt.imshow(moore, cmap='Purples')
```


Then, we establish the rules governing our cellular automata model. 
- If a cell has n infected neighbours in a Moore neighbourhood, and probability of getting Covid-19 from one neighbour is pi, the cell has (1-(1-pi)^n) probability of being infected. 
- If a cell is infected, the cell becomes recovered at day 10. All infected cells are assumed to recover. 
- The recovered state in our universe grid is represented by the number 3. By creating a separate state, we ensure a recovered individual does not return to a susceptible state, therefore they are assumed to have infinite immunity.

Having ran initial model, we added an additional state to simulate health policy intervention in the pandemic development- self isolation.

- If the cell is infected, there is a probability of pq of going into self isolation on each day after the incubation period. 
- New state of the quarantined cell means it is no longer infectious to its Moore neighbourhood. In contrast, an unquarantined cell remains infectious until recovery.


# Methodology


The figure below illustrates the overall pipeline of our model.

The rules described above are implemented iteratively, with each iteration signifying a "new day".

Firstly, to generate the start of the infection, a small proportion of individuals are scattered throughout our grid. Using random binomial generator, we generated a 13 by 13 grid filled with 0s and 1s, which represent not infected and infected individual respectively. The number of infected individuals depend on the parameter initial infection rate. This method effectively randomly scatters infected individuals amongst our population, simulating the spread of Covid-19 with multiple epicentres. The edges are 'ring-fenced' with 0s to avoid issues with counting the number of neighbours in edge cases.

For an uninfected individual (which has state 0), it is said to be 'exposed' if there is at least 1 neighbour in its neighbourhood that has been infected (show state 1). The probability of catching the disease increases exponentially with the number of infected neighbours in its neighbourhood. 

An infected individual (which has state 1) has a certain probability of going into quarantine (i.e. self isolation) starting from day 4, until recovery. This is implemented using random binomial generator, equivalent to a  Bernoulli trial with 2 outcomes, quarantined (probability pq) and not quarantined (probability 1-pq). We have chosen day 4 to be the start as it is in line with the average global incubation period of Covid-19 (Paul and Lorin, 2021). Effectively, we are using a uniform distribution to model the probability of a person going into quarantine on a given day. This variation in duration of self-isolation reflects the nature of Covid-19 pandemic as there may be delay in getting the test results, and the incubation period varies from one individual to another. 

Once an infected cell has been ill for 10 days, we alter the state of the cell to "3", which means recovered and no longer infectious to its neighbours.


For our self-isolation and recovery rules to work, we have to record the time each cell has been infected. Thus, we have created a time matrix, the same dimension as our population matrix. Each cell in the time matrix records the number of days the corresponding cell has been infected.


```python
#time matrix records the duration a cell has been infected
def count_time(input_uni):
    for i in range(input_uni.shape[0]):
        for j in range(input_uni.shape[1]):
            if input_uni[i,j] >0:
                #if a cell is infected
                if generation(input_uni,output_uni)[i,j] ==1 or generation(input_uni, output_uni)[i,j]==2:
                    #and it is infected the next day
                    time[i,j]+=1
```

```python
def generation(input_uni, output_uni):
    for i in range(input_uni.shape[0]):
        for j in range(input_uni.shape[1]):
            output_uni=G3.infectrecovery(input_uni, output_uni, i, j, p_inf,t)
            output_uni=G3.quarantine(t, output_uni,output_uni, i,j, pq,incubation_period)
    return(output_uni)
```

We have also created counting functions to count the number of susceptible, infected, recovered individuals. All individuals (not on the edge) who have not been infected are considered susceptible. These are stored in lists, which updates with each iteration of our rules. 

```python
#Newly infected cases
dc=[]
def num_infected(time):
    dc.append(np.sum(time==1))
```

```python
cr=[]
def cumulative_recovered(time):
    cr.append(np.sum(time==10))
```

```python
sus=[]
def susceptible(init):
    sus.append(np.sum(init==0)-(4*init.shape[0]-4))
#the outermost cell are ringfenced with 0s and thus we take them out of the susceptible population.
```

```python
def initialise():
    global dc
    global cr
    global sus
    dc=[]
    cr=[]
    sus=[]
    return(dc,cr,sus)
```

```python
def generate_imshow(timestep):
    if i%timestep==0:
        fig=plt.figure()
        plt.imshow(init, cmap='Purples')
        plt.clim(0,3)
        plt.colorbar()
        plt.title("infection status on the "+str(i)+"th day")
#generating colourmaps using imshow
```

# Results #

"Covid-19" spreads rapidly after initial random seeding of cells (fig. 1a). On average, 5% of cells are infected at the start.

Cells are colour coded to match their corresponding state: susceptible, infected, quarantined or recovered (0,1,2,3 respectively, from white to dark purple).
The probability of catching Covid-19 from an infected neighbour was to set  0.1, and probability of self-isolation from day 4 onwards set to 0.2. Within 16 days, all cells are infected.



```python
init=G3.get_universe(15,0.05)
time=np.zeros((15,15))
p_inf=0.1
pq=0.2
t=time
incubation_period=4

initialise()

for i in range(30):
    # Add a snapshot of the universe, then move to the next generation
    output_uni=np.copy(init)
    count_time(init)
    num_infected(time)
    cumulative_recovered(time)
    susceptible(init)
    generate_imshow(5)
    init=generation(init,output_uni)
```

## The effect of vaccination on the number of cases

We have added another parameter, vaccination rate, to our model to simulate the spread of Covid-19. A probabilistic vaccination matrix, such as the one above, is generated every time using random binomial generator. We used similar functions to those listed previously, with the addition of the states - vaccinated and unvaccinated, which correspond to different rates of contracting the disease.

```python
def infectrecoveryvac(input_uni, output_uni, x,y,pi,pivac,time):
    if input_uni[x, y] == 0:
        numnei=G3.neighbour(input_uni,x,y)
        if vacmat[x,y]==0:  #chances of catching Covid-19 if unvaccinated
            pnotvac = 1-((1-pi)**numnei)
            output_uni[x, y] = random.binomial(1, pnotvac, size=None) 
        else:  #chances of catching Covid-19 if unvaccinated
            pvac = 1-((1-pivac)**numnei)
            output_uni[x, y]=random.binomial(1, pvac, size=None)
    elif input_uni[x,y]==1 or input_uni[x,y]==2:
        if time[x,y]==10: #all infected cells recover on day 10
            output_uni[x,y] = 3
        else:
            output_uni[x,y]=input_uni[x,y]
    return(output_uni)
```

# Discussion

Overall our results model a rapid spread of disease in the 'neighbourhood' and slow propagation of quarantine. Perhaps the most important takeaway has been the effectiveness of this quarantine. Because the susceptible population has decreased by the time infected individuals are going into self isolation, our results would indicate that a quarantine implemented early is essential to protect even a small population of susceptible individuals. Increasing the infection rate in our model further contracted the time period over which the cases occurred (fig. 2). In the real world, this reflects the importance of a timely quarantining procedure to reduce the strain on healthcare systems.

The infectiousness of Covid-19 has increased over the last two years with the emergence of different variants, the most notorious being Delta and Omicron. We hypothesise that continuing increases in infection rate will produce even sharper and steeper rises in daily infection rates, as demonstrated in figure 3, relative to the strain of Covid-19 first identified in 2019.

From fig. 4 we can see that high vaccination rates correspond to a greatly reduced initial spike in daily cases and no sharp increases later on. This corresponds to the real-world concept of "smoothing the peak", or "flattening the curve". Vaccination is shown to be essential for preventing overload of healthcare systems during daily case spike events, which are more severe in the 50%/no vaccine conditions.

Finally, figure 2 demonstrated the effect of changes to quarantine rates. Infected individuals may not experience any of the symptoms associated with the disease, therefore mass quarantine measures are again shown to be essential in drastically reducing the number of daily infected cases. Such "lockdown" measures have been deployed globally over the last two years, and continue to be used in certain countries. However, with many countries having high vaccination rates against Covid-19, both public opinion and government policy are shifting away from mass quarantine methods, for instance the U.K. government's ruling out of further lockdowns.

### Consideration of limitations
In our development of this model several potential limitations, relating especially to its representability have become apparent.

Firstly, we have an extremely small population. We have designed it so it simulates the initial spread of an infection as well as saves computational power. In future, we could expand the grid to have more individuals.

We also have a static population. This is relatively true for a population in strict lockdown, however, this is not representative of the current situation, where people are allowed to travel and gather. We could simulate the movement of people using the random.shuffle, which shuffles each column in our matrix. The number of infected people remain the same before and after shuffling (see Fig 5a and 5b).


```python
universe=G3.get_universe(10,0.1)
fig=plt.figure()
plt.imshow(universe, cmap='Purples')
plt.title("Fig 5a.Pre-shuffling")
np.random.shuffle(universe[1:8,1:8])
fig=plt.figure()
plt.imshow(universe, cmap='Purples')
plt.title("Fig 5b. Post-shuffling")
```
